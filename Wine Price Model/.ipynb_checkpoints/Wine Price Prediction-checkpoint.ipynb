```python
# PREPROCESSING 
df = df[(df['price'] <= 60) & (df['price'] > 20)]
df['country'].fillna('Unknown', inplace=True)
df['region_1'].fillna('Unknown', inplace=True)
df['wine_type'] = map_varieties(df['variety']) # Map the grape varieties to wine types
```

```python
features = df[['points', 'country', 'region_1', 'wine_type']] # Creating the feature set and target
target = df['price']
```

```python
# One-hot ENCODING
column_transformer = ColumnTransformer(
    [("ohe", OneHotEncoder(handle_unknown='ignore'), ['country', 'region_1', 'wine_type'])],
    remainder='passthrough'
)
```

```python
# MODEL CREATION
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
model_pipeline = make_pipeline(column_transformer, Ridge()) # Create and train the model
model_pipeline.fit(X_train, y_train)
```

```python
# Function to make price prediction
def model_prediction(country, region, score, grape_variety):
    wine_type = map_varieties(grape_variety)
    input_data = pd.DataFrame([[score, country, region, wine_type]], columns=['points', 'country', 'region_1', 'wine_type'])
    price_prediction = model_pipeline.predict(input_data)
    return price_prediction[0], wine_type.strip().lower()
```

```python
# Create a GUI
window = tk.Tk()
window.geometry("300x250")
window.title("Wine Price Prediction")
```

```python
# Define a function to update the dropdown values as the user types
def update_dropdown(event):
    # Delay the update slightly after a key is released
    window.after(3000, do_update)

def do_update():
    current_text = grape_var.get()
    filtered_values = [value for value in df['variety'].unique() if current_text.lower() in value.lower()]
    grape_dropdown['values'] = filtered_values
    if len(filtered_values) > 0:
        grape_dropdown.event_generate("<Down>")
    else:
        grape_dropdown.event_generate("<Escape>")
```

```python
def create_box_plot(predicted_price, rmse, grape_variety, region, country, wine_color, score):
    # Create a synthetic dataset representing the range
    lower_bound = max(predicted_price - rmse, 0)  # Ensure lower bound is not negative
    upper_bound = predicted_price + rmse
    synthetic_data = np.random.normal(loc=predicted_price, scale=rmse/3, size=1000)
    synthetic_data = np.clip(synthetic_data, lower_bound, upper_bound)  # Clip values to the bounds

    # Create a horizontal box plot
    plt.figure(figsize=(12, 5))
    plt.boxplot(
        synthetic_data, 
        vert=False,  # Horizontal box plot
        patch_artist=True,  # Fill with color
        boxprops=dict(facecolor=wine_color, color=wine_color),
        medianprops=dict(color='black'),  # Median line color
        whiskerprops=dict(color=wine_color),
        capprops=dict(color=wine_color)
    )

    # Customize the plot
    plt.yticks([1], [f"{grape_variety.title()}\nfrom\n{region},\n{country}"])
    plt.xlabel(f"Predicted Price: ${predicted_price:.2f}")
    plt.xlim([lower_bound, upper_bound])  # Set the x-axis to show the range
    plt.title(f"Price Range for a {score}-point {grape_variety.title()} from {country}\n(Model RMSE: ${rmse:.2f})")
    
    # Display the plot
    plt.show()
```

```python
def calculate_rmse(m, X_test, y_test):
    y_pred = m.predict(X_test)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    return rmse
```

```python
# Define a function to predict the price and update the result label with a range bar
def features_to_predict():
    # Get user input from widgets
    country = country_var.get()
    region = region_var.get()
    score = score_var.get()
    grape_variety = grape_var.get().strip().lower()

    # Check if the score is within the valid range (50 to 100)
    if score < 50 or score > 100:
        result_label.config(text="Invalid point score")
    else:
        # Call the predict_price function
        predicted_price, wine_type = model_prediction(country, region, score, grape_variety)
        rmse = calculate_rmse(model_pipeline, X_test, y_test)
        predicted_price = float(predicted_price)

        # Calculate the lower and upper bounds based on RMSE
        lower_bound = predicted_price - rmse
        upper_bound = predicted_price + rmse

        # Determine the color based on wine type (red, gold, or pink)
        if wine_type == 'white':
            wine_color = 'gold'
        elif wine_type == 'rose':
            wine_color = 'pink'
        elif wine_type == 'red': 
            wine_color = 'purple'
        else: wine_color = 'black'

        create_box_plot(predicted_price, rmse, grape_variety, region, country, wine_color, score)

        # Update the result label with the predicted price and RMSE (Optional)
        result_label.config(text=f"Predicted Price: ${predicted_price:.2f}\nRMSE: {rmse:.2f}")
```

```python
# Dropdown for Country
country_var = tk.StringVar(window)
country_var.set(df['country'].unique()[0])  # default value
country_label = tk.Label(window, text="Country")
country_label.pack()
country_dropdown = ttk.Combobox(window, textvariable=country_var)
country_dropdown['values'] = sorted(list(df['country'].unique()))
country_dropdown.pack()
```

```python
# Function to update regions based on selected country
def update_regions(*args):
    selected_country = country_var.get()
    regions = sorted(list(df[df['country'] == selected_country]['region_1'].dropna().unique()))
    region_dropdown['values'] = regions
    if len(regions) > 0:
        region_var.set(regions[0])
    else:
        region_var.set('Unknown')

# Bind the country dropdown selection event to update regions
country_var.trace('w', update_regions)
```

```python
# Dropdown for Region
region_var = tk.StringVar(window)
region_label = tk.Label(window, text="Region")
region_label.pack()
region_dropdown = ttk.Combobox(window, textvariable=region_var)
region_dropdown.pack()

# Initialize regions based on the default selected country
update_regions()
```

```python
# Dropdown for Grape
grape_var = tk.StringVar(window)
grape_var.set(df['variety'].unique()[0])  # first value
grape_label = tk.Label(window, text="Grape Variety")
grape_label.pack()
grape_dropdown = ttk.Combobox(window, textvariable=grape_var)
grape_dropdown['values'] = sorted(list(df['variety'].unique()))
grape_dropdown.pack()

# Bind events to Grape Combobox for autocomplete
grape_dropdown.bind("<KeyRelease>", lambda event: update_dropdown(event))
```

```python
# Validation function for score entry
def validate_score(P):
    if P == "" or P.isdigit():
        if P == "" or 50 <= int(P) <= 100:
            return True
    return False

# Validation command
vcmd = (window.register(validate_score), '%P')
```

```python
# Entry for Score with validation
score_var = tk.DoubleVar(window)
score_label = tk.Label(window, text="Score")
score_label.pack()
score_entry = tk.Entry(window, textvariable=score_var, validate='key', validatecommand=vcmd)
score_entry.pack()
```

```python
# Button to predict price
predict_button = tk.Button(window, text="Predict Price", command=features_to_predict)
predict_button.pack()
```

```python
# Label for showing the result
result_label = tk.Label(window, text="")
result_label.pack()
```
